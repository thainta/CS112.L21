### **1.ABSTRACTION (TRỪU TƯỢNG HÓA)**
- Tóm tắt bài toán:

 + Ta có 2 phân số tối giản: $ \frac{a}{b} $ và $ \frac{c}{d} $.

 + Phương pháp biến đổi: Mỗi lần biến đổi tăng $a$, $b$ lên $1$ đơn vị, sau đó tối giản phân số $\frac{a}{b}$

 + Xác định số bước biến đổi để $ \frac{a}{b} = \frac{c}{d} $, nếu không tìm thấy thì **Số bước = 0** 


- Abstraction: Tìm số bước biến đổi **incr** sao cho $ \frac{a}{b} = \frac{c}{d} $. Nếu không tìm thấy thì in ra **0**

### **2. PATTERN RECOGNIZATION (NHẬN DIỆN KHUÔN MẪU)**
- Kỹ thuật áp dụng: **Duyệt** 


### **3. ALGORITHM DESIGNED (THIẾT KẾ THUẬT TOÁN)**
#### **3.1 Mã giả:**
* **`incr = 0`** (Với **incr** là số bước biến đổi)
* Trong khi **`(a/b) < (c/d)`**:
  + Tăng **`incr = incr + 1`**
  + Biến đổi phân số : 
     + Biến đổi tử: **`a = a + 1`**
     + Biến đổi mẫu: **` b = b + 1`**
  + Nếu **`(a/b) = (c/d)`**:
    + Output: Xuất kết quả **incr**
    + Thoát chương trình
* Output: Xuất **0** (Không tìm thấy)

#### **3.2 Đánh giá độ phức tạp:**
  - Độ phức tạp trong bài này là: **O(2 * max(b, d))**

#### **3.3 Code:**
 

```python
import math
a = int(input())
b = int(input())
c = int(input())
d = int(input())
result = 0
if a / b > c / d :
    print(0)
else:
    while(a / b < c / d):
      a+=1
      b+=1
      x = math.gcd(a, b)
      a//=x
      b//=x
      if a / b > c / d:
        result = 0
        break 
      result+=1
    print(result)
```
